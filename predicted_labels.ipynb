```python
from ultralytics import YOLO

model = YOLO("/home/umang.shikarvar/instaformer/delhi.pt") # Load the YOLO model

image_dir = "/home/umang.shikarvar/instaformer/delhi_CUT_bks/images" # Directory containing images for prediction

# Run prediction and save labels

results = model.predict(
    source=image_dir,
    save_txt=True,  # Saves labels as .txt files
    conf=0.001,  # Confidence threshold (0.001 = 0.1%)
    imgsz=640,
    iou=0.33,  # NMS IoU threshold 
    save_conf=False,  # Save confidence scores in labels
    max_det=300,  # Maximum detections per image
)
```
